```python
import numpy as np
import scipy as sp



from scipy.stats import norm


import matplotlib.pyplot as plt 
```

```python
def visual(result):
    
    
    plt.plot(result)
```

```python
def normal_distribution(x,mean,std):
    
    result1=f"The probability distribution function is {norm.pdf(x,mean,std)  }"   ##probability density function 
    
    result2=f"The cumulative distribution functions is { norm.cdf(x,mean,std)}"
    
    
    result3=f"The percent point function is {norm.ppf(x,mean,std)}"
    
    
    
    
    
    for i in [result1,result2,result3]:
        
        
        visual(i)
        
        
        
    
    return result1,result2,result3
```

```python
x=np.linspace(1,100,1000)
mean=34
std=45

normal_distribution(x,mean,std)

```

```python
#  there are 100 students in the class and in one of the mathematics tests the average marks scored by the students in the subject is 78 and the standard deviation is 25. The marks of the student follow Normal probability distribution. We can use this information to answer some questions about the students marks. 

# Python Code for Percentage of Students who got less than 60 marks
# Here we will use the norm() function from scipy.stats module to make the probability distribution for the populations mean equal to 78 and the standard deviation equal to 25. 
```

```python

x=60

mean=78
std=25


# z_score is calculated when we have to calculate less than or greater than of particular value 

z_score=(x-mean)/std


probability=norm.cdf(z_score)



prob_percentage=probability*100


prob_percentage
```
